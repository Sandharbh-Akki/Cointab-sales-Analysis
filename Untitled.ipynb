```python
awb_data=comp.groupby('AWB Code').agg({
    'Order ID': 'first',
    'Charged Weight': 'first',
    'Warehouse Pincode_x': 'first',
    'Customer Pincode': 'first',
    'Type of Shipment': 'first',
    'Billing Amount (Rs.)': 'first',
    'Zone_y': 'first',
    'Weight Slabs': 'first',
    'Forward Fixed Charge': 'first',
    'Forward Additional Weight Slab Charge': 'first',
    'RTO Fixed Charge': 'first',
    'RTO Additional Weight Slab Charge': 'first',
    'SKU': 'count',
    'Order Qty': 'sum',
    'Actual weight(kg)': 'sum'
})
```

```python
awb_data.head(1)
```

```python
awb_data['Slab_count']=np.ceil(awb_data['Actual weight(kg)']/awb_data['Weight Slabs'])
```

```python
awb_data.columns
```

```python
awb_data
```

```python
def actual_price(x):
    if x['Type of Shipment'] == 'Forward charges':
        return x['Slab_count'] * x['Forward Additional Weight Slab Charge'] + x['Forward Fixed Charge']
    elif x['Type of Shipment'] == 'Forward and RTO Charges':
        return (x['Slab_count'] * x['Forward Additional Weight Slab Charge'] + x['Forward Fixed Charge']) + \
               (x['Slab_count'] * x['RTO Additional Weight Slab Charge'] + x['RTO Fixed Charge'])
    else:
        return 'error'
```

```python
awb_data['actual price']=awb_data.apply(lambda x:  actual_price(x),axis=1)
```

```python
awb_data
```

```python
# Check data types
print(awb_data.dtypes)

# Convert columns to numeric
awb_data['actual price'] = pd.to_numeric(awb_data['actual price'], errors='coerce')
awb_data['Billing Amount (Rs.)'] = pd.to_numeric(awb_data['Billing Amount (Rs.)'], errors='coerce')

# Check if conversion was successful
print(awb_data.dtypes)

# Perform subtraction
awb_data['actual vs invoice'] = awb_data['actual price'] - awb_data['Billing Amount (Rs.)']
```

```python
awb_data
```

```python
awb_data
```

```python
def finds(x):
    if x['actual vs invoice'] < 0:
        return 'under charged'
    elif x['actual vs invoice'] == 0:
        return 'correctly charged'
    else:
        return 'over charged'
```

```python
awb_data['Status']=awb_data.apply(lambda x: finds(x),axis=1)
```
